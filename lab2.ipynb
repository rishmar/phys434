```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
X = 2
Y = 5
```

```python
dist = stats.poisson(X)
x = np.arange(0, 11)
pmf = dist.pmf(x)
plt.title("Poisson distribution for cosmic background")
plt.xlabel("Average cosmic rays")
plt.ylabel("Probability")
plt.vlines(x, 0, pmf, linewidth = 20)
plt.xticks(x)
plt.ylim(0, None);
```
