```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
X = 2
Y = 5
```

```python
dist = stats.poisson(X)
x = np.arange(0, 11)
pmf = dist.pmf(x)
plt.title("Poisson distribution for cosmic background")
plt.xlabel("Observed average cosmic rays")
plt.ylabel("Probability")
plt.bar(x, pmf, linewidth = 5)
```

```python
sum_p = np.convolve(pmf, pmf)
plt.title("Summed probability distribution after 2 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.bar(range(len(sum_p)), sum_p, linewidth = 5)
plt.yscale('log')
```

```python
pmf2 = pmf
for j in range(1, 5):
    pmf2 = np.convolve(pmf2, pmf)
plt.bar(range(len(pmf2)), pmf2, linewidth = 5)
plt.title("Summed probability distribution after 5 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.yscale('log')
```
