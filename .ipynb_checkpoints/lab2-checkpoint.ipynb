# Problem 1

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
X = 2
Y = 5
```

```python
dist = stats.poisson(X)
x = np.arange(0, 11)
pmf = dist.pmf(x)
plt.title("Poisson distribution for cosmic background")
plt.xlabel("Observed average cosmic rays")
plt.ylabel("Probability")
plt.bar(x, pmf, linewidth = 5)
```

```python
sum_p = np.convolve(pmf, pmf)
plt.title("Summed probability distribution after 2 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.bar(range(len(sum_p)), sum_p, linewidth = 5)
plt.yscale('log')
```

```python
pmf2 = pmf
for j in range(1, 5):
    pmf2 = np.convolve(pmf2, pmf)
plt.bar(range(len(pmf2)), pmf2, linewidth = 5)
plt.title("Summed probability distribution after 5 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.yscale('log')
```

```python
# The average part c
pmf = dist.pmf(x)
pmf3 = pmf
for i in range(1, 10):
    pmf3 = np.convolve(pmf3, pmf)
plt.vlines(np.arange(len(pmf3)) / 11, 0, pmf3)
plt.title("Average probability distribution after 10 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.yscale('log')
plt.xlim(0, 21)
plt.ylim(1e-10, 1);
```

```python
# What will the graph looks like after 50 days?
pmf = dist.pmf(x)
pmf3 = pmf
for i in range(1, 20):
    pmf3 = np.convolve(pmf3, pmf)
plt.vlines(np.arange(len(pmf3)) / 21, 0, pmf3)
plt.title("Average probability distribution after 20 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.yscale('log')
plt.xlim(0, 21)
plt.ylim(1e-10, 1);
```

```python
# What will the graph looks like after 100 days?
pmf = dist.pmf(x)
pmf3 = pmf
for i in range(1, 30):
    pmf3 = np.convolve(pmf3, pmf)
plt.vlines(np.arange(len(pmf3)) / 31, 0, pmf3)
plt.title("Average probability distribution after 30 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.yscale('log')
plt.xlim(0, 1)
plt.ylim(1e-10, 1);
```

# Problem 2

```python
dist = stats.expon(scale = 1)
x = np.linspace(0, 5, 1000)
plt.plot(x, dist.pdf(x), linewidth = 5)
plt.title("Probability density function of exponential distribution")
plt.ylabel("Probability density")
plt.xlabel("x")
```

```python
pdf = dist.pdf(x)
new_pdf = pdf
for i in range(10):
    new_pdf = np.convolve(new_pdf, pdf)
plt.plot(np.linspace(0, 5, len(new_pdf)), new_pdf)
plt.title("Pdf after 10 intervals")
plt.xlabel("Average x")
plt.ylabel("Probability density")
plt.yscale('log')
plt.xlim(0, 5)
plt.ylim(1e-50, None);
```

```python
pdf = dist.pdf(x)
new_pdf = pdf
for i in range(50):
    new_pdf = np.convolve(new_pdf, pdf)
plt.plot(np.linspace(0, 5, len(new_pdf)), new_pdf)
plt.title("Pdf after 50 intervals")
plt.xlabel("Average x")
plt.ylabel("Probability density")
plt.yscale('log')
plt.xlim(0, 5)
plt.ylim(1e-50, None);
```

```python
pdf = dist.pdf(x)
new_pdf = pdf
for i in range(100):
    new_pdf = np.convolve(new_pdf, pdf)
plt.plot(np.linspace(0, 5, len(new_pdf)), new_pdf)
plt.title("Pdf after 100 intervals")
plt.xlabel("Average x")
plt.ylabel("Probability density")
plt.yscale('log')
plt.xlim(0, 5)
plt.ylim(1e-50, None);
```
