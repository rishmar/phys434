```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
import pandas as pd
from scipy import stats
plt.rcParams["figure.figsize"] = (10,8)
```

## Problem 0

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

## Problem 1

1a) Make a scatter plot (not a line plot) of the first 1000 data points.

```python
x = t[0:1000]
y = bkgd[0:1000]
plt.scatter(x, y)
plt.title('The scatter plot for first 1000 data points')
plt.ylabel('Background Data')
plt.xlabel('Time')
plt.show()
```

1b) Generalize your code so you can make a plot of any X contiguous points (e.g. 2000 points starting at point 500), and produce an example plot of a set of data somethere in the middle of your array.

```python
def makescatter(start, background, time, step = 1):
    x = time[start:start + step]
    y = background[start:start + step]
    plt.scatter(x,y)
    plt.title("The scatter plot of " + str(background)+" from "+ str(start) + "data points from ")
    plt.ylabel('Background Data')
    plt.xlabel('Time')
    plt.show()
```

```python
makescatter(2000, bkgd, t, 2000)
```

1c) Sometimes you want to sample the data, such as plotting every 100th point. Make a plot of the full data range, but only every 100th point.

```python
def scatter100(t, background, step):
    plt.scatter(t[::step],background[::step])
    plt.title("The scatter plot of every 100th data points ")
    plt.ylabel('Background Data')
    plt.xlabel('Time')
    plt.show()
```

```python
scatter100(t, bkgd, 100)
```

## Problem 2

2a) Make a 2d histogram plot. 

```python
plt.hist2d(t, bkgd, density=True, bins=150)
plt.title('2D hisgraph of Background Datas')
plt.xlabel('Time')
plt.ylabel('Background')
plt.colorbar()
plt.show()
```

2b) Clearly explain what is being plotted in your plot 

The above plot shows the density of datas and we colorized each bins based on how many they have. The more yellow a bin is, the greater density of data points. The more violet a bin is, means the less density of datas. Look through the graph, the yellow area is near to the center of the plot. at the top and botton is getting more violet.
