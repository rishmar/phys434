2. State the statistical question in words. Be very clear, and describe why you have chosen this statistical question.
Restate your question in math. Be very clear.

Suppose there's a temperature measurement X. What is the probability that a measurement at 12 Kelvin within the SD 0.4, would be far away from mean than X? Suppose measurement cannot exceeds 3 $\sigma$.

3. Restate your question in math. Be very clear.

probability of X from gaussian distribution with 12Kelvin within SD 0.4: $$P(X) = 2 \int_{|X-12|}^\infty\frac{1}{0.4}e^{\frac{-1}{2} (\frac{x}{0.4})^2}$$

Then we use quantile function: $$Q(p) = (2p-1) \sqrt{\frac{1}{2p-2p^2}}$$

Finally, if the answear is bigger than 3, we discard the answear.

4. Apply your statistical test and construct a truth table.

```python
dist = abs(stats.norm.rvs(loc = 12., scale = 0.4, size = 100000) - 12)
bad = abs(np.array(bad) - 12)
good_kept = 0
good_throw = 0
bad_kept = 0
bad_throw = 0

p1 = 1 - stats.norm.cdf(-dist, scale = 0.4)
ppf1 = stats.norm.ppf(p1, loc = 0, scale = 1)
p2 = 1 - stats.norm.cdf(-bad, scale = 0.4)
ppf2 = stats.norm.ppf(p2,loc = 0, scale = 1)

# go through each ppf.
for x1 in ppf1:
    if x1 <= 3:
        good_kept +=1
    else:
        good_throw +=1
        
        
for x2 in ppf2:
    if x2 <= 3:
        bad_kept +=1
    else:
        bad_throw +=1
        
print("Good kept: ", good_kept, "\nGood throw: ", good_throw)
print("Bad kept: ", bad_kept, "\nBad throw: ", bad_throw)
```

## Problem 2

In this example we will be looking for asteroids. If we look at the alignment of stars on subsequent images, they don't perfectly align due to atmospheric and instrumental effects (even ignoring proper motion). The resulting distribution is two dimensional, and for this lab let's assume it is a 2D Gaussian with 1 arcsecond RMS. Or said another way, if I histogram how far all the (stationary) stars appear to have moved I get something like:

```python
a = np.vstack((stats.norm.rvs( scale = 1, size = 100000), stats.norm.rvs( scale = 1, size = 100000)))
a.shape
fig, ax = plt.subplots(1, 1)
h = ax.hist2d(a[0,:],a[1,:],bins=100, density=True);
ax.set_aspect('equal', 'box')
plt.xlim([-3 , 3])
plt.ylim([-3 , 3])
plt.title("2D Histogram of positional uncertainty", fontsize = 24)
plt.ylabel("$\Delta$y arcseconds", fontsize = 18)
plt.xlabel("$\Delta$x arcseconds", fontsize = 18)
plt.colorbar(h[3], ax=ax)
```

**We would like a '5 sigma' detection of movement. What is that distance in arcseconds?**

*1. Question in words*

What is the radius around the center of the image that detections outside of that radius have more than or equal to $5\sigma$?

*2. Question in math*

We knoe that the magnitude of a vector of two independent Gaussians is a Rayleigh distribution, the pdf of this Rayleigh distribution is: $p(x)=\frac{x}{\sigma^2}e^{\frac{-x^2}{2\sigma^2}}$

Gaussian distributions with $\sigma = 1$, we got the cdf: $ C(r) = 1-e^{\frac{-a^2}{2}}$ and Quantile function is: $Q=\sqrt{-2\ln(1-C(r))}$ 

If the answear of Q is larger than 5, then it's discard.

*3. What distance in arcseconds represents a 5 sigma detection of motion?*

```python
# A while loop to increase the distance until we reach 5 sigma.
d = 0
sigma = 0
n = 0

while n < 4: 
    p = stats.rayleigh.cdf(d)
    sigma = stats.norm.ppf(p)
    if(sigma < 5):
        d = d + 0.1**n
    else:
        d = d - 0.1**n
        n += 1
n -= 1
d = d + 0.1**n
sigma
```

$\lambda = 7200$ and $k = 6800$
